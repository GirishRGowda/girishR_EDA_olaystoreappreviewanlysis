## Taking Care of Null values

```python
# checking the number of null values presented in each column
df.isna().sum()
```

The median ratings of each relevant "Category" are used to fill in the null values of "Rating" in order to increase the outcomes of our study.

```python
# to fill with 'Category' median, we are using groupby and transform method
df["Rating"] = df.groupby("Category")['Rating'].transform(lambda x: x.fillna(x.median()))
```

Now replacing the null values in the "Type," "Current Ver," and "Android Ver" columns with the corresponding modes.

```python
# filling mode of themselves in null values of columns using fillna method
df['Type'].fillna(df['Type'].mode()[0], inplace=True) # we used [0] because we are taking value at 0th index of the mode() output
df['Current Ver'].fillna(df['Current Ver'].mode()[0], inplace=True)
df['Android Ver'].fillna(df['Android Ver'].mode()[0], inplace=True)
```

```python
df.info()
```

as seen in info, the following column's Data Type is "Object."



*  Price
*  Review
*  Installs

we must convert it to their preferred integer and float types.

```python
# for Price
# 1. removing the '$' sign in 'Price' column
df['Price'] = df['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
# 2. converting the string type price to float type 
df['Price'] = df['Price'].apply(lambda x: float(x))


# for Review, we use pandas 'to neumeric' method cause it has large numbers
df['Reviews'] = pd.to_numeric(df['Reviews'],errors='coerce')


# for Installs, 
# 1. we first replace unwanted characters from numeric string
df['Installs'] = df['Installs'].str.replace('+','').str.replace(',','').str.replace('-','')
# 2. now we convert it to integer using apply method
df['Installs'] = df['Installs'].apply(lambda x: int(x))
```

For best performance, we transform the data's string-formatted, MB or KB values from the "Size" column to MB.

```python
df['Size']= df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x) #     Replacing 'M' from string of Size
df['Size']= df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
df['Size']= df['Size'].apply(lambda x: float(str(x).replace('k', ''))/1000 if 'k' in str(x) else x) #converting Kb of size to MB by dividing to 1000
df['Size']= df['Size'].apply(lambda x: float(x)) # converting all values to float

df['Size'].fillna(float(int(df['Size'].mean())), inplace=True) # filled nan values of size with the mean of size
df
```

# Analysing another collection of user review data.

```python
# getting the dataset from google drive

df1 = pd.read_csv('/content/drive/MyDrive/User Reviews.csv')

df1.info()
```

### Currently, a few important columns have a high percentage of null values, so we are removing them since they contain no information.

```python
# we are merging the previous data set and the new one with App column, since the new data set has additional information of reviews in it
new_df = df.merge(df1, on='App')

# few columns of new data set has null values, so dropping them as they didn't have any information
new_df.dropna(axis=0,subset=['Translated_Review', 'Sentiment','Sentiment_Polarity','Sentiment_Subjectivity'], inplace=True)

new_df.head()
```

```python
new_df.info()
```

## The number of positive, negative, and neutral emotions in each category is analysed.

```python
plt.figure(figsize=(15, 15))

# Get the count of each sentiment by category
tdf = new_df.loc[:,['Category','Sentiment']]
tdf = tdf.groupby(['Category', 'Sentiment']).size().reset_index(name='counts')

# Pivot the dataframe to get the sentiment as columns
tdf = tdf.pivot(index='Category', columns='Sentiment', values='counts')

# Plot the stacked bar chart
tdf.plot(kind='barh', stacked=True)
plt.legend(title='Sentiment')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

Insights:-

Overall, positive sentiments are higher than negative and neutral sentiments in reviews, with gaming having the highest positive and negative sentiments at about 60% and 40%, respectively. All other categories also follow this pattern, with positive reviews being higher than negative reviews and neutral reviews being the lowest, leading us to the conclusion that no category is performing particularly negatively.

## Calculating the positive sentiment percentage for each category from high to low

```python
# counting categoies of ony positive sentiments and assign to new dataframe
tdf1 = pd.DataFrame(new_df[new_df['Sentiment']=='Positive']['Category'].value_counts().sort_values(ascending=False)).reset_index()

# counting categoies of ony Negetive sentiments and assign to new dataframe
tdf2 = pd.DataFrame(new_df[new_df['Sentiment']=='Negative']['Category'].value_counts().sort_values(ascending=False)).reset_index()

# counting categoies of ony Negetive sentiments and assign to new dataframe
tdf3 = pd.DataFrame(new_df[new_df['Sentiment']=='Neutral']['Category'].value_counts().sort_values(ascending=False)).reset_index()

tdf = tdf1.merge(tdf2, on='index') 
tdf = tdf.merge(tdf3, on='index')

tdf.columns=['Category','Positive','Negative','Neutral'] 

# adding new column in temp data frame of percentage where we store percentage of positive sentiments of each category
tdf['Positive percentage'] = round((tdf['Positive']/(tdf['Positive']+tdf['Negative']+tdf['Neutral']))*100,2)

tdf = tdf.sort_values(['Positive percentage'],ascending=False) 
```

```python
plt.figure(figsize=(15,8))

# Plot the positive percentage for each category
plt.plot(tdf['Category'], tdf['Positive percentage'],'r-o')

# Customize the plot
plt.title('Positive Sentiment Percentage for Each Category', fontsize=20)
plt.xlabel('Category')
plt.ylabel('Positive Percentage')
plt.xticks(rotation=90, fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

Insights:-

The graphical representation shows that 5 categories have more than 75% of the total reviews that are favourable, indicating that these categories have already satisfied their customers. In contrast, the categories with lower levels of positivity have not yet achieved this level of customer satisfaction.
